```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['font.family'] = 'NanumGothic'
plt.rcParams['axes.unicode_minus'] = False
```

```python
df = pd.read_csv('/content/drive/MyDrive/2. 멋사/amazon_bestsellers.csv')

df.info()
```

```python
df.head()
```

```python
plt.figure(figsize=(15, 6))
plt.subplot(1, 2, 1)
sns.countplot(x='Genre', data=df, palette='pastel') # 장르(Genre)별로 개수를 세어 막대 그래프를 그립니다.
plt.title('장르별 베스트셀러 도서 수', fontsize=15) # 그래프 제목
plt.xlabel('장르', fontsize=12)
plt.ylabel('도서 수', fontsize=12)
```

```python
plt.subplot(1, 2, 2)
# 사용자 평점(User Rating)의 분포를 히스토그램으로 그립니다.
sns.histplot(df['User Rating'], bins=10, kde=True, color='skyblue')
plt.title('사용자 평점 분포', fontsize=15) # 그래프 제목
plt.xlabel('사용자 평점', fontsize=12)
plt.ylabel('빈도', fontsize=12)
```

```python
# 상위 10명 작가를 출력합니다.
top_10_authors = df['Author'].value_counts().head(10)
print("\n상위 10명의 작가:")
print(top_10_authors)
```
